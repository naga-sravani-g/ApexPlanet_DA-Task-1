**2. Initial Data Inspection**

```python
import pandas as pd

df = pd.read_csv("amazon_sales_dataset.csv")
df.head()
```

```python
data_dictionary.to_csv("data_dictionary.csv", index=False)
```

**4. Box Plot Visulization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=df["price"])
plt.title("Box Plot of Product Price")
plt.xlabel("Product Price")
plt.show()
```

**5. Count Plot Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(y=df['product_category'], order = df['product_category'].value_counts().index)
plt.title('Distribution of Product Categories')
plt.xlabel('Count')
plt.ylabel('Product Category')
plt.show()
```

This count plot shows the number of occurrences for each `product_category`. The categories are ordered from most frequent to least frequent.

**6. Duplicate Handling**

```python
df = df.drop_duplicates()
```

**7. Missing Value Imputation**

```python
df.fillna("Unknown", inplace=True)  # categorical
```

**8. Data Colume Conversion**

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

**9. Feature Engineering**

```python
from datetime import datetime

df["Order_Age_Years"] = datetime.now().year - df["order_date"].dt.year
```

**10. Text Standardization**

```python
df["product_category"] = df["product_category"].str.lower().str.strip()
```

**11. Cleaned Data Export**

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
